**Introduction**:  
Sleep is essential for overall health, impacting physical and mental well-being. Lifestyle factors such as stress, physical activity, and blood pressure can significantly influence sleep patterns. High stress levels and low physical activity often lead to poor sleep quality and duration, contributing to sleep disorders. Elevated blood pressure and unhealthy sleep habits can further exacerbate these issues, underscoring the importance of a balanced lifestyle for healthy sleep.

**Features Description** :
The dataset consists of the following columns:

* Person ID: An identifier for each individual in the dataset.

* Gender: The gender of the person (Male/Female).

* Age: The age of the person in years.

* Occupation: The occupation or profession of the person.

* Sleep Duration (hours): The number of hours the person sleeps per day.

* Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10

* Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.

* Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.

* BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).

* Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.

* Heart Rate (bpm): The resting heart rate of the person in beats per minute.

* Daily Steps: The number of steps the person takes per day.

* Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
data = pd.read_excel("/content/df_final_new_2.xlsx")
```

```python
#small x, CAPS Y
from sklearn.linear_model import LogisticRegression

x1= data.drop(columns=["Person ID","Sleep Disorder"])
y1= data["Sleep Disorder"]

numeric_features = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps', 'Systolic', 'Diastolic']
categorical_features = ['Gender', 'Occupation', 'BMI Category']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

x_train,x_test,Y_train,Y_test = train_test_split(x1,y1,test_size = 0.2, random_state=42)

x_train = preprocessor.fit_transform(x_train)
x_test=preprocessor.transform(x_test)

model_lr = LogisticRegression(max_iter = 100, random_state=42)
model_lr.fit(x_train,Y_train)

pred_y_lr = model_lr.predict(x_test)

print(classification_report(Y_test,pred_y_lr))
```

```python
#ONE HOT ENCODED FEATURES (small x CAPS Y)
print(Y_test)
print(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_dtc = DecisionTreeClassifier(random_state=42)
model_dtc.fit(x_train,Y_train)

pred_y_dtc = model_dtc.predict(x_test)

print(classification_report(Y_test,pred_y_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rfc = RandomForestClassifier(random_state=42)
model_rfc.fit(x_train,Y_train)

pred_y_rf = model_rfc.predict(x_test)

print(classification_report(Y_test,pred_y_rf))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model_gbc = GradientBoostingClassifier(random_state=42)
model_gbc.fit(x_train,Y_train)

pred_y_gbc = model_gbc.predict(x_test)

print(classification_report(Y_test,pred_y_gbc))
```

```python
from sklearn.ensemble import AdaBoostClassifier

model_abc = AdaBoostClassifier(random_state=42)
model_abc.fit(x_train,Y_train)

pred_y_abc = model_abc.predict(x_test)

print(classification_report(Y_test,pred_y_abc))
```

```python
from sklearn.ensemble import BaggingClassifier

model_bag = BaggingClassifier(random_state=42)
model_bag.fit(x_train,Y_train)

pred_y_bag = model_bag.predict(x_test)

print(classification_report(Y_test,pred_y_bag))
```

```python
from sklearn.svm import SVC

model_svc = SVC(random_state=42)
model_svc.fit(x_train,Y_train)

pred_y_svc = model_svc.predict(x_test)

print(classification_report(Y_test,pred_y_svc))
```

```python
def predict_sleep_disorder(model, input_data):
    """
    Predicts the sleep disorder based on input data.

    Parameters:
    model: Trained model
    input_data: Dictionary containing input features

    Returns:
    Prediction: 0 (Healthy), 1(Insomnia) , or 2 (Sleep apnea)
    """
    input_df = pd.DataFrame([input_data])
    prediction = model.predict(input_df)
    return prediction[0]

#Using XGB model
input_data = {
    'Gender': 1,
    'Age': 30,
    'Occupation': 2,
    'Sleep Duration': 7.0,
    'Quality of Sleep': 5,
    'Physical Activity Level': 50,
    'Stress Level': 6,
    'BMI Category': 1,
    'Heart Rate': 70,
    'Daily Steps': 8000,
    'Systolic': 120,
    'Diastolic': 80
}

predicted_disorder = predict_sleep_disorder(model_xgb, input_data)
if predicted_disorder == 0:
    print("Predicted Sleep Disorder: Healthy")
elif predicted_disorder == 1:
    print("Predicted Sleep Disorder: Insomnia")
elif predicted_disorder == 2:
    print("Predicted Sleep Disorder: Sleep Apnea")

print(predicted_disorder)
```

# Confusion Matrix and ROC - AUC

Random Forest and XGBoost models have achieved high accuracy of 90% and therefore, below are their Confusion Matrices and ROC - AUC of these models

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc


cm_rf = confusion_matrix(y_test, pred_y_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve
y_prob_rf = model_rf.predict_proba(X_test)
fpr_rf = {}
tpr_rf = {}
roc_auc_rf = {}

for i in range(3):
    fpr_rf[i], tpr_rf[i], _ = roc_curve(y_test, y_prob_rf[:, i], pos_label=i)
    roc_auc_rf[i] = auc(fpr_rf[i], tpr_rf[i])

plt.figure()
for i in range(3):
    plt.plot(fpr_rf[i], tpr_rf[i], label=f'ROC curve of class {i} (area = {roc_auc_rf[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC Curve')
plt.legend(loc="lower right")
plt.show()

# ROC AUC Score
roc_auc_rf = roc_auc_score(y_test, y_prob_rf, multi_class='ovr')
print(f"Random Forest ROC AUC Score: {roc_auc_rf:.2f}\n")
```
